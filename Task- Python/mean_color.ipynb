```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
buf = plt.imread("img.jpg")
plt.imshow(buf)
buf.shape
```

```python
buf = buf/255
plt.imshow(buf)
```

```python
mn = np.mean(buf,axis=(0,1))
mn
```

```python
to_show = [[mn]]
```

```python
plt.imshow(to_show)
```

```python
def show_main_color(path):
    fig=plt.figure(figsize=(20, 6))
    fig.add_subplot(1, 2, 1)
    plt.imshow(plt.imread(path))
    plt.axis('off')
    fig.add_subplot(1, 2, 2)
    plt.imshow([[np.mean((plt.imread(path)/255),axis=(0,1))]])
    plt.axis('off')
    plt.show()
```

```python
show_main_color('img.jpg')
```
