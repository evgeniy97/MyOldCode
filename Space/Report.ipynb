<center> **Теория** <center>
    $\frac{dx}{dt} = f(x)$  
    $f(x_0)=f_0$
<center> Представим производные, как разностную схему: <center>
    $\frac{x_{n+1}-x_{n}}{\delta t} = f(x_n)$  
    $x_{n+1} = x_n + \delta t f(x_n)$
 <center> Тем самым мы получили метод Эйлера первого порядка <center>

Для рассматриваемой задачи:  

$\frac{dx}{dt}=v$  
$\frac{dv}{dt}=G\frac{M}{r^3}r$
$x(0)=x_0,v(0)=v_0$  
G берем равное 1. Тогда примения разностные схемы и метод Эйлера получаем:  
$x_{n+1} = x_n + v_{n-1} \delta t + G\frac{M}{r^3}r (\delta t)^2  $


Часть кода, отвечающая за реализацию метода Эйлера:  

<code>def Euler(self):
        r = sqrt((self.x - X0)**2 + (self.y - Y0)**2)
